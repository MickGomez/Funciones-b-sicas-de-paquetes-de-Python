# Matplotlib -> libreria para graficos de python
## Sirve para:
* Gráficos 2D

## Gráfica de puntos o lineas:
* Visualizar una cantidad considerable de datos
* Buena para estudiar tendencias entre variables

```python
# importar pandas
import pandas as pd
# datos RRSS 2020
fb = ['Facebook', 2449, True,2006]
tw = ['Twitter', 339, False,2006]
ig = ['Instagram',1000, True,2010]
yt = ['Youtube',2000,False,2005]
lk = ['LinkedIn',663,False,2003]
wa = ['Whatsapp',1600,True,2009]
lista_rrss = [fb,tw,ig,yt,lk,wa]
# crear DataFrame a partir de listas
df_rrss = pd.DataFrame(lista_rrss, 
             columns=['Nombre','Cantidad','Pertenece a Facebook','Año'])
df_rrss
```

```python
# Importar el paquete Matplotlib 
import matplotlib.pyplot as plt 
```

```python
# ejemplo básico
x=[1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.show()
```

## Utilizando el DataFrame de las redes sociales

```python
# grafica de lineas utilizando matplotlib
plt.plot(df_rrss['Nombre'],df_rrss['Cantidad'])
plt.show()
```

```python
# grafica de puntos 
plt.scatter(df_rrss['Nombre'],df_rrss['Cantidad'])
plt.show()
```

```python
# Hacer lo mismo utilizando metodos del DataFrame de pandas
df_rrss.plot(kind='scatter',x='Nombre',y='Cantidad')
plt.show()
```

# Gráficas de barras
* Buena para conteos
* Agrupa datos por resumen
* Fácil para explicar

```python
# graficas de barras
plt.bar(df_rrss['Nombre'],df_rrss['Cantidad'])
plt.show()
```

```python
# graficas de barras ordenada
df_rrss_sort = df_rrss.sort_values('Cantidad', ascending=False)
plt.bar(df_rrss_sort['Nombre'],df_rrss_sort['Cantidad'])
plt.show()
```

```python
# graficas de barras ordenada y con color 
plt.bar(df_rrss_sort['Nombre'],df_rrss_sort['Cantidad'],
       color=['b','r','g','m','b','k'] )
plt.show()
# b = blue
# g = green
# r = red
# c = cyan
# m= magenta
# y = yellow
# b = black
# w = white
```

# Gráfica de Pastel/Torta/Pie
* No siempre es recomendable
* Proporciones como ángulos
* No hacer en 3D

```python
# grafica de torta/pie
plt.pie(df_rrss['Cantidad'],
       labels=df_rrss['Nombre'])
plt.show()
```

```python
# utilizar colores 
lista_col_rrss = ['#4267B2','#FF0000','#25D366','#E1306C','#006699','#1DA1F2']
```

```python
# grafico de pie
plt.pie(df_rrss_sort['Cantidad'], 
       labels=df_rrss_sort['Nombre'],
       colors=lista_col_rrss)
plt.show()
```
